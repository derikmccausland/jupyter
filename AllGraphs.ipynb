# Importing and Cleaning Data
Here I am using a combination of Pandas, Python, and Google Sheets to clean and orient data in the best way to suit my needs.

```python
#Importing a few altered .csv files
allPok = pd.read_csv('all-clean.csv')
attSplit = pd.read_csv('attribute.csv');
```

```python
#Grouping Dataframes for use in Bar Graphs
typeHealth = pd.DataFrame(data = allPok.groupby(['type','year'])['hp'].max())
typeHealth.reset_index(inplace = True);

attMax = pd.DataFrame(data = attSplit.groupby(['type','year','Attribute'])['value'].max())
attMax.reset_index(inplace = True);
```

```python
# Health Calculations
meanh = pd.DataFrame(data = allPok.groupby('year')['hp'].mean())
meanh.reset_index(inplace = True)
maxh = pd.DataFrame(data = allPok.groupby('year')['hp'].max())
maxh.reset_index(inplace = True)
maxh.rename(columns={"hp": "max-hp"}, inplace=True)
extracted_col = meanh["hp"]
maxh = maxh.join(extracted_col)
maxh.rename(columns={"hp": "mean-hp"}, inplace=True)

# Damage Calculations
meand = pd.DataFrame(data = allPok.groupby('year')['damage'].mean())
meand.reset_index(inplace = True)
maxd = pd.DataFrame(data = allPok.groupby('year')['damage'].max())
maxd.reset_index(inplace = True)
maxd.rename(columns={"damage": "max-d"}, inplace=True)
extracted_col = meand["damage"]
maxd = maxd.join(extracted_col)
maxd.rename(columns={"damage": "mean-d"}, inplace=True)
extracted_col = typeHealth["type"]
maxd = maxd.join(extracted_col)
```

```python
maxD = pd.DataFrame(data = allPok.groupby(['type','year'])['damage'].max())
maxD.reset_index(inplace = True)
damageRange = pd.read_csv('alright.csv')
damageRange1 = damageRange.drop(['year'], axis=1)
healthRange = pd.read_csv('hTime.csv')
healthRange1 = healthRange.drop(['year'], axis=1);
```

```python
allTime = pd.read_csv('allTime.csv')
allTime1 = healthRange.drop(['year'], axis=1);
```


# How often was a type the highest?
I wanted to see how often each type was the highest Damage or Health card for that year, let's look in some Pie Charts.


```python
#Behind the Scenes

pieTime = pd.DataFrame(damageRange['type'].value_counts())
pieTime.reset_index(inplace = True)
pieTimeH = pd.DataFrame(healthRange['type'].value_counts())
pieTimeH.reset_index(inplace = True);
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 8))
colors = sns.color_palette('muted')[0:5]
pieTime["both"] = pieTime['index'] +" - "+ pieTime['type'].astype(str)
pieTimeH["both"] = pieTimeH['index'] +" - "+ pieTimeH['type'].astype(str)
axs[0].pie(pieTime['type'], labels = pieTime['both'], colors = colors, autopct='%.0f%%')
axs[0].set_title('Number of Years Each Type had the Highest Damage')
axs[1].pie(pieTimeH['type'], labels = pieTimeH['both'], colors = colors, autopct='%.0f%%')
axs[1].set_title('Number of Years Each Type had the Highest Health');
```
