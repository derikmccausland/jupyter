```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
meanh = pd.DataFrame(data = allPok.groupby('year')['hp'].mean())
meanh.reset_index(inplace = True)
maxh = pd.DataFrame(data = allPok.groupby('year')['hp'].max())
maxh.reset_index(inplace = True)
maxh.rename(columns={"hp": "max-hp"}, inplace=True)
extracted_col = meanh["hp"]
maxh = maxh.join(extracted_col)
maxh.rename(columns={"hp": "mean-hp"}, inplace=True)
```

```python
damageRange = pd.read_csv('alright.csv')
damageRange1 = damageRange.drop(['year'], axis=1)
pieTime = pd.DataFrame(damageRange['type'].value_counts())
pieTime.reset_index(inplace = True)
```

```python
sns.set(rc = {'figure.figsize':(20,8)})

ax = damageRange1.set_index('type').plot(kind='bar', stacked=True)

plt.xlabel('Type')
plt.ylabel('Damage Potential')

plt.title('Highest Damage Potentail Per Year')
for i in ax.containers:
    ax.bar_label(i,)
    
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

change_width(ax, .75);
```

```python
colors = sns.color_palette('pastel')[0:5]
pieTime["both"] = pieTime['index'] +" - "+ pieTime['type'].astype(str)
#create pie chart
plt.pie(pieTime['type'], labels = pieTime['both'], colors = colors, autopct='%.0f%%')
plt.title('Number of times each type had the highest damage for that year')
plt.show()
```
