```python
damageRange = pd.read_csv('alright.csv')
damageRange1 = damageRange.drop(['year'], axis=1)
pieTime = pd.DataFrame(damageRange['type'].value_counts())
pieTime.reset_index(inplace = True)
```

```python
sns.set(rc = {'figure.figsize':(20,8)})

ax = damageRange.set_index('year').plot(kind='bar', stacked=True)

plt.xlabel('Year and Highest Card Type')
plt.ylabel('Damage Breakdown')

plt.title('Highest Damage Occurrences by Type')
for i in ax.containers:
    ax.bar_label(i,)
    
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

change_width(ax, .75);
```

```python
colors = sns.color_palette('pastel')[0:5]
pieTime["both"] = pieTime['index'] +" - "+ pieTime['type'].astype(str)
#create pie chart
plt.pie(pieTime['type'], labels = pieTime['both'], colors = colors, autopct='%.0f%%')
plt.title('Number of Years Each Type had the Highest Damage')
plt.show()
```
